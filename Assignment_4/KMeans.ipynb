```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Import data
data = pd.read_csv('Shopping.csv')
data.head()
```

```python
# Plot points
plt.figure(figsize = (7, 4))
plt.scatter(data.iloc[:, 1], data.iloc[:, 2])
plt.xlabel('Spending Index')
plt.ylabel('Income Index')
plt.show()
```

```python
# Convert pandas dataframe to numpy ndarray
data_array = np.array(data[['SpendingIndex', 'IncomeIndex']])
```

```python
# KMeans clustering
def kmeans_clustering(n_clusters):
    model = KMeans(n_clusters = n_clusters)
    model_fit = model.fit(data_array)
    return model_fit.labels_, model_fit.cluster_centers_
```

```python
# Get cluster labels for k = 2,3 and 4
labels_2, centers_2 = kmeans_clustering(2)
labels_3, centers_3 = kmeans_clustering(3)
labels_4, centers_4 = kmeans_clustering(4)

# Plot clusters
fig, ax  = plt.subplots(3, 1, figsize = (7, 15))

ax[0].scatter(data.iloc[:, 1], data.iloc[:, 2], c = labels_2, cmap = 'rainbow')
ax[0].scatter(centers_2[:, 0], centers_2[:, 1], marker = 'o', s = 100)
ax[0].set_title('K = 2')

ax[1].scatter(data.iloc[:, 1], data.iloc[:, 2], c = labels_3, cmap = 'rainbow')
ax[1].scatter(centers_3[:, 0], centers_3[:, 1], marker = 'o', s = 100)
ax[1].set_title('K = 3')

ax[2].scatter(data.iloc[:, 1], data.iloc[:, 2], c = labels_4, cmap = 'rainbow')
ax[2].scatter(centers_4[:, 0], centers_4[:, 1], marker = 'o', s = 100)
ax[2].set_title('K = 4')
```

From the clusters above, 
1. K = 2 seggregates the points well as expected visually. 
2. With K = 3, we see that shoppers with high income index and moderate-high shopping index are grouped together.
3. At K = 4, the clustering seems most appropriate with the following behaviour clustered together:

    1. shoppers with low income index and high shopping index
    2. shoppers with moderate income index and high shopping index
    3. shoppers with high income index and moderate shopping index
    4. shoppers with moderate income index and low shoppiing index
